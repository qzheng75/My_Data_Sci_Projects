## Feature matching

Extract defining key features from an input image, then use distance calculation to find all matchs in another image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display(img, cmap='gray'):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
reeses = cv2.imread('../DATA/reeses_puffs.png', 0)
display(reeses)
```

```python
cereals = cv2.imread('../DATA/many_cereals.jpg', 0)
display(cereals)
```

## Brute-force detection using ORB descriptors

```python
orb = cv2.ORB_create()
kp1, des1 = orb.detectAndCompute(reeses, None)
kp2, des2 = orb.detectAndCompute(cereals, None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(des1, des2)
```

```python
matches = sorted(matches, key=lambda x : x.distance)
```

```python
reeses_matchs = cv2.drawMatches(reeses, kp1, cereals, kp2, matches[:25], None, flags=2)
```

```python
display(reeses_matchs)
```

## Scale-Invariant Feature Transform

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp1, des1 = sift.detectAndCompute(reeses, None)
kp2, des2 = sift.detectAndCompute(cereals, None)
```

```python
bf = cv2.BFMatcher()
# k: number of best matches to find
matches = bf.knnMatch(des1, des2, k=2)
```

```python
good_match = []
for match1, match2 in matchs:
    if match1.distance < 0.75 * match2.distance:
        good_match.append([match1])
```

```python
len(good_match), len(matches)
```

```python
sift_matches = cv2.drawMatchesKnn(reeses, kp1, cereals, kp2, good_match, None, flags=2)
```

```python
display(sift_matches)
```

## FLANN

```python
sift = cv2.xfeatures2d.SIFT_create()
kp1, des1 = sift.detectAndCompute(reeses, None)
kp2, des2 = sift.detectAndCompute(cereals, None)
```

```python
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm=FLANN_INDEX_KDTREE, trees=5)
search_params = dict(checks=50)
```

```python
flann = cv2.FlannBasedMatcher(index_params, search_params)
```

```python
matches = flann.knnMatch(des1, des2, k=2)
np.array(matches).shape
```

```python
matchesMask = [[0, 0] for i in range(len(matches))]
```

```python
good = []
for match1, match2 in matches:
    if match1.distance < 0.7 * match2.distance:
        good.append([match1])
```

```python
len(good), len(matches)
```

```python
flann_matches = cv2.drawMatchesKnn(reeses, kp1, cereals, kp2, good, None, flags=2)
display(flann_matches)
```

```python
good = []
for i, (match1, match2) in enumerate(matches):
    if match1.distance < 0.7 * match2.distance:
        matchesMask[i] = [1, 0]
```

```python
draw_params = dict(matchColor=(0, 255, 0),
                   singlePointColor=(255, 0, 0),
                   matchesMask=matchesMask,
                   flags=2)
```

```python
flann_matches = cv2.drawMatchesKnn(reeses, kp1, cereals, kp2, matches, None, **draw_params)
display(flann_matches)
```
