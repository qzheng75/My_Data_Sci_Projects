```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_img():
    img = np.zeros((600, 600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img, text='ABCDE', org=(50, 300),
               fontFace=font, fontScale=5, color=(255, 255, 255),
               thickness=25)
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = load_img()
display_img(img)
```

```python
kernel = np.ones((5, 5), dtype=np.uint8)
kernel
```

```python
# Erosion
result = cv2.erode(img, kernel, iterations=5)
display_img(result)
```

```python
img = load_img()
white_noise = np.random.randint(low=0, high=2, size=(600, 600))
display_img(white_noise)
```

```python
# Set to same scale of the original image
white_noise = white_noise * 255
```

```python
noise_img = white_noise + img
display_img(noise_img)
```

```python
# Opening: Erosion + dilation
# can be used to clear the background noise of the image
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel)
display_img(opening)
```

```python
# Create some foreground noise
img = load_img()
black_noise = np.random.randint(low=0, high=2, size=(600, 600))
black_noise *= -255
black_noise_img = black_noise + img
black_noise_img[black_noise_img == -255] = 0
display_img(black_noise_img)
```
