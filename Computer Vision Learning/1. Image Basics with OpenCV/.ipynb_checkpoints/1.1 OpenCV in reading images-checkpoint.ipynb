```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
# Passing wrong path: result in None
img = cv2.imread('../DATA/00-puppy.jpg')
type(img)
```

```python
plt.imshow(img)
```

```python
# matplotlib: RGB in the order
# opencv: BGR order
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread('../DATA/00-puppy.jpg',
                     cv2.IMREAD_GRAYSCALE)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
fix_img.shape
```

```python
resized_img = cv2.resize(fix_img, (1000, 400))
plt.imshow(resized_img)
```

```python
w_ratio = 0.5
h_ratio = 0.4 # 40% of the original
squeezed_img = cv2.resize(fix_img, (0, 0),
                          fix_img, w_ratio, h_ratio)
plt.imshow(squeezed_img)
plt.title(f'New shape: {squeezed_img.shape}')
```

```python
fliped_img = cv2.flip(fix_img, 0)
plt.imshow(fliped_img)
plt.title('Filped along vertical axis')
```

```python
fliped_img = cv2.flip(fix_img, 1)
plt.imshow(fliped_img)
plt.title('Filped along horizontal axis')
```

```python
# Save an image
cv2.imwrite('../DATA/new_puppy.jpg', fix_img)
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)
ax.imshow(fix_img)
```
