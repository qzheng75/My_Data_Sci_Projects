Line Charts

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Prepare data
x = np.linspace(0, 20, 1000)
y = np.sin(x) + 0.2 * x
z = np.log(x)

# Plot it
plt.plot(x, y, label='sinx + 0.2x', color='r')
plt.plot(x, z, label='log(x)', color='b')

# Add labels and a legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trigonometry')
plt.legend()

# Show the plot
plt.show()
```

Scatterplot

```python
# Create data
X = np.random.randn(100, 2)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# Make scatterplot with different markings
X = np.random.randn(200, 2)
X[:50] += 3
Y = np.zeros(200)

# Label first 50 dots as 1
Y[:50] = 1

# Plot with labels
plt.scatter(X[:, 0], X[:, 1], c=Y)
plt.show()
```

Histogram

```python
X = np.random.randn(10000)
plt.hist(X)
plt.show()
```

```python
# Plot with more bars
plt.hist(X, bins=40)
plt.show()
```

```python
# Plot XOR dataset
X0 = np.random.rand(2000)
X1 = np.random.rand(2000)
X = np.vstack((X0, X1)).T - 0.5

Y = np.zeros(2000)
Y[(X[:, 0] > 0) & (X[:, 1] > 0)] = 1
Y[(X[:, 0] < 0) & (X[:, 1] < 0)] = 1

plt.scatter(X[:, 0], X[:, 1], c=Y)
plt.show()
```
