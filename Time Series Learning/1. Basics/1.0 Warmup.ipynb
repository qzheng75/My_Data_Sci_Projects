Warmup 01:
- Use numpy to generate 1000 samples from the standard normal $X$ ~ $\mathcal{N}(0, 1)$
- Plot $X$ as a time series
- Plot $X$ as a histogram

```python
import numpy as np
import matplotlib.pyplot as plt
X = np.random.randn(1000)
```

```python
time_steps = range(len(X))
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].plot(time_steps, X)
axes[0].set_title('Time series')
axes[1].hist(X, bins=40)
axes[1].set_title('Histogram')
plt.show()
```

Warmup 02:
- Use numpy to generate 1000 samples for $X$
- Generate $y = 0.8x + 0.4$ with random noise
- Plot a scatterplot for $X, y$
- Find and plot the line of best fit

```python
from sklearn.linear_model import LinearRegression

X = np.linspace(0, 20, 1000)
noise = np.random.normal(0, 2, 1000)
y = 0.8 * X + 0.4 + noise
plt.scatter(X, y)

model = LinearRegression()
model.fit(X.reshape(-1, 1), y)

x_range = np.linspace(0, 20, 100)
y_pred = model.predict(x_range.reshape(-1, 1))  # Predict y values
plt.plot(x_range, y_pred, color='r', label='Line of Best Fit');
```

Warmup 03:
- Call `np.cumsum()` on the noise
- Plot the results

```python
noise_sum = np.cumsum(noise)
plt.plot(X, noise_sum);
```

Warmup 04:
- Generate and plot 1000 samples from multivariate normal: $X$ ~ $\mathcal{N}(0, 1)$ with:

\begin{align*}
\mu = \begin{bmatrix} 0 \\ 0 \end{bmatrix}, \Sigma = \begin{bmatrix} 1 & -0.5 \\ -0.5 & 2 \end{bmatrix}
\end{align*}

```python
mu = np.array([0, 0])
Sigma = np.array([[1, -0.5], [-0.5, 2]])

X = np.random.multivariate_normal(mu, Sigma, 1000)
plt.scatter(X[:, 0], X[:, 1]);
```
