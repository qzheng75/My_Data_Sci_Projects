```python
import numpy as np
import matplotlib.pyplot as plt

T = 1000 # time steps
P0 = 10 # Initial price
mu = 0.001 # Drift
```

```python
last_p = np.log(P0) # Last log price
log_returns = np.zeros(T)
prices = np.zeros(T)
for t in range(T):
    # sample a log return
    r = 0.01 * np.random.randn()
    p = last_p + mu + r
    
    log_returns[t] = r + mu
    prices[t] = np.exp(p)
    
    last_p = p
```

```python
plt.figure(figsize=(20, 8))
plt.plot(prices)
plt.title("Simulated Prices");
```

```python
# Vectorized implementation
log_returns = 0.01 * np.random.randn(T)
cumulative_log_returns = np.cumsum(log_returns)

# Compute prices
prices = P0 * np.exp(mu * np.arange(T) + cumulative_log_returns)
log_prices = np.log(prices)

plt.figure(figsize=(20, 8))
plt.plot(prices)
plt.title("Simulated Prices");
```
