### SMA
In time series processing, **SMA**, or Simple Moving Average, is a fundamental technique used to analyze and smooth out fluctuations in time series data. SMA calculates the average value of a specified number of data points within a moving window. By continuously updating this average over time, SMA provides a smoothed representation of the underlying trends and patterns in the data, making it easier to identify important insights and remove noise.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

close = pd.read_csv('sp500_close.csv', index_col=0, parse_dates=True)
```

```python
goog = close[['GOOG']].copy().dropna()
goog.head()
```

```python
goog.plot();
```

Remember formula for log return: $r_t = \log(1+R_t) = \log(1 + \frac{P_t}{P_{t-1}})$

```python
# Calculate log return
goog_ret = np.log(1 + goog.pct_change(1))
```

```python
goog_ret.plot();
```

```python
# SMA with sliding window size = 10
goog['SMA-10'] = goog['GOOG'].rolling(10).mean()
goog.head(15)
```

```python
goog.plot();
```

```python
# With window size = 50. The plot is smoother
goog['SMA-50'] = goog['GOOG'].rolling(50).mean()
goog.plot()
```

```python
# Do SMA on two variables
goog_aapl = close[['GOOG', 'AAPL']].copy().dropna()
cov = goog_aapl.rolling(50).cov()
cov
```

```python
# Get the covariance on a single day
cov.loc['2018-02-07'].to_numpy()
```

```python
# Calculate log return for two variables
goog_aapl_ret = np.log(1 + goog_aapl.pct_change(1))
goog_aapl_ret.head()
```

```python
goog_aapl_ret['GOOG-SMA-50'] = goog_aapl_ret['GOOG'].rolling(50).mean()
goog_aapl_ret['AAPL-SMA-50'] = goog_aapl_ret['AAPL'].rolling(50).mean()
goog_aapl_ret.plot(figsize=(10, 5));
```

```python
cov = goog_aapl_ret[['GOOG', 'AAPL']].rolling(50).cov()
cov.tail()
```

```python
# See correlation between two companies on each day
corr = goog_aapl_ret[['GOOG', 'AAPL']].rolling(50).corr()
corr.tail(16)
```
