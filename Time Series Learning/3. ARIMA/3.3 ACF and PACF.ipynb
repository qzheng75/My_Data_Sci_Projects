```python
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
import numpy as np
import matplotlib.pyplot as plt
```

### PACF

```python
# Plot PACF for random series
x0 = np.random.randn(1000)
plt.plot(x0);
```

```python
# Default confidence interval: alpha = 0.05
fig, ax = plt.subplots()
plot_pacf(x0, ax=ax);
```

```python
# Design a series which each term depends on previous term
# tau = 1 should be non-zero
x1 = [0]
for i in range(1000):
  x = 0.5 * x1[-1] + 0.1 * np.random.randn()
  x1.append(x)
x1 = np.array(x1)
plt.plot(x1);
```

```python
fig, ax = plt.subplots()
plot_pacf(x1, ax=ax);
```

```python
# Each term is negatively related to the previous term
x1 = [0]
for i in range(1000):
  x = -0.5 * x1[-1] + 0.1 * np.random.randn()
  x1.append(x)
x1 = np.array(x1)
plt.plot(x1);
```

```python
fig, ax = plt.subplots()
plot_pacf(x1, ax=ax);
```

```python
# Now each term is dependent on previous two terms
# tau = 1, 2 should be non-zero
x2 = [0, 0]
for i in range(1000):
  x = 0.5 * x2[-1] - 0.3 * x2[-2] + 0.1 * np.random.randn()
  x2.append(x)
x2 = np.array(x2)
plt.plot(x2);
```

```python
fig, ax = plt.subplots()
plot_pacf(x2, ax=ax);
```

```python
# Depend on 5 previous terms
x5 = [0, 0, 0, 0, 0]
for i in range(1000):
  x = 0.5 * x5[-1] - 0.3 * x5[-2] - 0.6 * x5[-5] + 0.1 * np.random.randn()
  x5.append(x)
x5 = np.array(x5)
plt.plot(x5);
```

```python
fig, ax = plt.subplots()
plot_pacf(x5, ax=ax);
```

### ACF

```python
# iid noise
fig, ax = plt.subplots()
plot_acf(np.random.randn(1000), ax=ax);
```

```python
# Depend on one term
errors = 0.1 * np.random.randn(1000)
ma1 = []
for i in range(1000):
  if i >= 1:
    x = 0.5 * errors[i-1] + errors[i]
  else:
    x = errors[i]
  ma1.append(x)
ma1 = np.array(ma1)
plt.plot(ma1);
```

```python
fig, ax = plt.subplots()
plot_acf(ma1, ax=ax);
```

```python
# 6 terms
errors = 0.1 * np.random.randn(1000)
ma6 = []
for i in range(1000):
  x = 0.5 * errors[i-1] - 0.3 * errors[i-2] + 0.7 * errors[i-3] + \
      0.2 * errors[i-4] - 0.8 * errors[i-5] - 0.9 * errors[i-6] + errors[i]
  ma6.append(x)
ma6 = np.array(ma6)
```

```python
fig, ax = plt.subplots()
plot_acf(ma6, ax=ax);
```
