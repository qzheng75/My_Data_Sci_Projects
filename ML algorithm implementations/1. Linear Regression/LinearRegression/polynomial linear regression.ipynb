```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from linear_regression import LinearRegressionModel, plot_linear_fit
```

```python
data = pd.read_csv('../data/non-linear-regression-x-y.csv')
x = data['x'].values.reshape((data.shape[0], 1))
y = data['y'].values.reshape((data.shape[0], 1))
data.head(10)
```

```python
plt.plot(x, y)
plt.show()
```

```python
num_iterations = 50000  
learning_rate = 0.02  
polynomial_degree = 15  
sinusoid_degree = 15  
normalize_data = True  

model = LinearRegressionModel(x, y, polynomial_degree, sinusoid_degree, normalize_data)
hist = model.train(num_iterations, learning_rate)
print(f"Initial cost: {hist[0]}, final cost: {hist[-1]}")
```

```python
params_table = pd.DataFrame({'Model Parameters': model.w.flatten()})
params_table.head(5)
```

```python
plt.plot(range(num_iterations), hist)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Gradient Descent Progress')
plt.show()
```

```python
predictions_num = 1000
x_predictions = np.linspace(x.min(), x.max(), predictions_num).reshape(predictions_num, 1);
y_predictions = model.predict(x_predictions)

plt.scatter(x, y, label='Training Dataset')
plt.plot(x_predictions, y_predictions, 'r', label='Prediction')
plt.show()
```
